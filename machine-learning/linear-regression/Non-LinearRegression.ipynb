```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from linear_regression_gradient_descent import LinearRegression
```

#### 通过设置ploynomial_degree和sinusoid_degree的值，有152个特征

```python
data = pd.read_csv('./data/non-linear-regression-x-y.csv')

x = data['x'].values.reshape(-1, 1)
y = data['y'].values.reshape(-1, 1)

num_iterations = 1000
learning_rate = 0.02
ploynomial_degree = 15
sinusoid_degree = 15
normalize_data = True

linear_regression = LinearRegression(x, y, ploynomial_degree, sinusoid_degree, normalize_data)

theta, cost_history = linear_regression.train(learning_rate, num_iterations)
# print(theta) # 需要时152维

print('开始的损失{:.2f}'.format(cost_history[0]))
print('结束的损失{:.2f}'.format(cost_history[-1]))
```

```python
plt.plot(range(num_iterations), cost_history)
plt.xlabel('iteration')
plt.ylabel('cost')
plt.title('Gradient Descent Progress')
plt.show()
```

```python
predictions_num = 1000
x_predictions = np.linspace(x.min(), y.max(), predictions_num).reshape(-1, 1)

predict_y = linear_regression.predict(x)
y_predictions = linear_regression.predict(x_predictions)
```

#### 过拟合

```python
plt.scatter(x, y, c='r', label='Training Dataset')
plt.scatter(x, predict_y, c='g', label='Prediction')
plt.scatter(x_predictions, y_predictions, c='b', label='Prediction')
plt.title('contrast')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
from matplotlib.image import imread


image = imread('ladybug.png')

X = image.reshape(-1, 3)
n_colors = (10, 8, 6, 4, 2)
segment_imgs = []
for n_cluster in n_colors:
    kmeans = KMeans(n_clusters=n_cluster, n_init=10, random_state=6).fit(X)
    segment_imgs.append(kmeans.cluster_centers_[kmeans.labels_].reshape(image.shape))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.subplot(2, 3, 1)
plt.imshow(image)
plt.title('initial image')

for index, n_color in enumerate(n_colors):
    plt.subplot(2, 3, 2 + index)
    plt.imshow(segment_imgs[index])
    plt.title('{} color'.format(n_color))

plt.show()
```

### 栈和队列的定义和特点

- 栈和队列是*使用频率最高的*数据结构
- 栈和队列是指插入和删除只能在表的**一端**进行的**线性表**
- **从数据结构角度看**，栈和队列*也是线性表*
    - 特殊性在于：栈和队列的基本操作是线性表操作的**子集**，是操作受限的线性表
        - 可称为限定性的数据结构
    - 栈：只能在**末尾插入和删除**
    - 队列：只能在**末尾插入**，在**表头删除**


**栈**——**先进后出**（限定性操作）
- 进制转换
- 括号匹配
- 表达式求值
- 函数调用
- 递归调用
- 迷宫的求解

**队列**——**先进先出**
- 脱机打印操作
- 进程按等待时间创建优先级队列
- 信号按接收的先后顺序依次处理

### 栈

栈（stack）：
- 限制在表的一端进行插入和删除运算的线性表

栈顶（top）和栈底（bottom）：
- 允许插入、删除的一端为栈顶，另一端是栈底
- 空栈：表中没有元素称为空栈

栈的运算规则：
- 先进后出 Last In First Out(LIFO)
- 添加元素：入栈、压栈push
- 删除元素：出栈pop


与一般线性表的**区别**：仅仅是运算规则不同
- 逻辑结构：一对一
- 存储结构：顺序表、链式表
- 运算规则：一般的是可以在**任意位置**插入和删除

    - 使用`append`和`pop`实现入栈和出栈

```
        self._top = self._top.next
        return p  # 返回栈顶元素
```

- 子程序（函数等）的调用和返回处理，在编译系统中采用**栈**来实现
- 当在一个**主程序**调用一个**子程序**时，在运行该主程序**前**需先将子程序之前的所有中间结果、返回地址等保存在**栈**中，子程序调用结束后将栈中的数据取出
- 多个子程序嵌套调用的规则是:**后调用-先返回**，内存管理实行“**栈式管理**”

### 队列

**队列**
- 运算受限的线性表，只允许在表的一端进行插入，而在另一端进行删除

**队头**（front）和**队尾**（rear）
- 允许删除的一端为对头，允许插入的一端为队尾

队列的运算规则
- **先进后出**（FIFO）



#### 顺序队列

**队列的顺序存储结构**
- 用*连续的存储空间*存放当前队列中的元素
- 另设**两个‘指针’**分别指示对头和队尾元素在队列中的位置
- 通常顺序存储的队列采用**循环队列结构**



**队头和队尾指针的位置和操作**：
- 对头指针front：指向队头元素所在的位置
    - 初始为空时，front为0
    - 出队时，取出队头元素，然后将队头指针加1（并返回被取出的元素）
- 队尾指针rear：指向队尾元素的*下一个*位置
    - 初始为空时，rear为0
    - 入队时，先将新元素插入到该位置，再将队尾指针加1

顺序队列结构**存在问题**
- 设顺序表容量为M，则：
    - 当front=0, rear=M-1时，再有元素入队发生溢出——真溢出
    - 当front≠0, rear=M-1时，再有元素入队发生溢出——**假溢出**

**循环队列**（circular queue）
- 将队列首尾连接成环形进行使用，让`sq[0]`接在`sq[M-1]`之后
    - 若`rear+1==M`，则令`rear=0`

**运算**：
```
if i + 1 == M:
    i = 0
else:
    i += 1
```
- 利用模运算（`%`），可以简化为
    - 入队：`rear = (rear + 1) % M`
    - 出队：`front = (front + 1) % M`



**循环队列中队满、队空判定条件**
- **问题**：
    - 队满：`front == rear`
    - 对空：`front == rear`
- **解决**：
    - **少用一个元素空间**
        - 队空：`front == rear`
        - 队满：`(rear + 1) % M == front`

#### 使用顺序栈实现十进制转换为其他进制[链接文字]
- 顺序栈在Python中是**用list实现的**

````python
class sstack():
    def __init__(self):
        self.slist = []

    def is_empty(self):
        if self.slist == []:
            return True
        else:
            return False

    def pushstack(self, data):
        self.slist.append(data)

    def popstack(self):
        return self.slist.pop()

    def peekdata(self):  # 取栈顶元素
        return self.slist[-1]

    def sizestack(self):
        return len(self.slist)

if __name__ == '__main__':
    N = int(input('输入一个十进制数：'))
    r = int(input('转换的进制：'))
    s1 = sstack()
    while N != 0:
        s1.pushstack(N % r)
        N = N // r
    while not s1.is_empty():
        data = s1.popstack()
        print(data, end=' ')
    print()

````

#### 链栈实现进制转换

````python
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None

class ListStack:
    def __init__(self):
        self._top = None

    def is_empty(self):
        return self._top

    def pushstack(self, data):
        node = Node(data)
        node.next = self._top
        self._top = node

    def popstack(self):
        p = self._top
        self._top = p.next
        return p  # 不用判断是否为空，在这里也能体现出来

if __name__ == '__main__':
    N = int(input('输入一个十进制数：'))
    r = int(input('转换的进制：'))
    s1 = ListStack()
    while N != 0:
        s1.pushstack(N % r)
        N = N // r
    while s1.is_empty():
        p = s1.popstack()
        print(p.data, end=' ')  # 只打印数据域
    print()
````

#### 链队列的实现

````python
class QueHeadNode:
    def __init__(self):
        self.front = None
        self.rear = None

class QueDataNode:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkQueue:
    def __init__(self):
        self._head = QueHeadNode()

    def is_empty(self):
        if self._head.front == self._head.rear and self._head.front is None:
            return True
        else:
            return False

    def enqueue(self, data):
        s = QueDataNode(data)
        if self.is_empty():
            self._head.front = s
            self._head.rear = s
        else:
            self._head.rear.next = s
            self._head.rear = s

    def dequeue(self):
        if self.is_empty():
            return
        if self._head.front == self._head.rear:
            p = self._head.front
            self._head.front = None
            self._head.rear = None
            return p
        else:
            p = self._head.front
            self._head.front = self._head.front.next
            return p

if __name__ == '__main__':
    que = LinkQueue()
    n = int(input('输入元素个数'))
    for i in range(n):
        data = int(input('输入元素：'))
        que.enqueue(data)
    while not que.is_empty():
        p = que.dequeue()
        print(p.data, end = ' ')
    print()
````
